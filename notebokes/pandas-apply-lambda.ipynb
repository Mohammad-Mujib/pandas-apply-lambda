```python
import pandas as pd
```

```python
df = pd.read_csv('../data/input/IMDB-Movie-Data.csv')
```

```python
df
```

```python
def movie_votes(votes):
    if 0 < votes <= 999 :
        return 1
    elif 999 < votes <= 9999 :
        return 2
    elif 9999 < votes <= 99999 :
        return 3
    elif 99999 < votes <= 999999 :
        return 4
    elif 999999 < votes :
        return
        
```

```python
df['bins'] = df.apply(lambda x : movie_votes(x['Votes']), axis=1)
```

```python
df.head()
```

```python
#2
df['Rev/min'] = df['Revenue (Millions)']/df['Runtime (Minutes)']
```

```python
df.head()
```

```python
new_ranking = lambda genre: 1 if genre == 'thriller' else -1 if genre == 'comedy' else 0
```

```python
df['new_ranking'] = df['Genre'].apply(new_ranking)
```

```python
#5
director_stats = df.groupby('Director').agg({'Rating': 'mean', 'Revenue (Millions)': 'mean'})
```

```python
director_stats['score'] = director_stats['Rating'] * director_stats['Revenue (Millions)']
```

```python
director_stats['ranking'] = director_stats['score'].rank(ascending=False)
```
